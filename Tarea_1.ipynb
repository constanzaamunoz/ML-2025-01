---

**Population:** Indica la medida de la población en el país. A mayor población, el GDP per cápita del país disminuye, mientras que no existan cambios en la producción. \
**Area (sq. mi.):** Un país con una mayor superficie (país grande) podría tener más recursos naturales (minería, agricultura, etc.) que un país más pequeño, por lo que su GDP podría ser mayor que el del país de menor superficie.\
**Pop. Density (per sq. mi.):**Esta variable indica la densidad poblacional. Un país con alta densidad poblacional puede generar economías de escala, ya que la concentración de personas facilita el desarrollo de infraestructura, servicios y negocios. Esto puede aumentar la productividad laboral y por consecuencia el GDP per capita. Sin embargo, una densidad extremadamente alta podría generar congestión, aumentar los costos de vida y disminuir la calidad de vida, afectando negativamente el GDP per capita.\
**Coastline (coast/area ratio):** A mayor Coastline, mayor podría ser el GDP per cápita al poder tener a disposición más puertos y por ende, mayor comercio (explicado en el ejemplo). \
**Net migration:** Esta variable indica la migración neta. De ser alta podría afectar al GDP per cápita de manera positiva, si consideramos que la inmigración es valiosa y aumenta la productividad del país, o de manera negativa, en caso de que la inmigración no sea valiosa y aumente la población sin aumentar la producción del país, logrando la disminución del GDP per cápita.\
**Infant mortality (per 1000 births):** Esta variable indica la tasa de mortalidad infantil. Una menor tasa indica una buena calidad de vida, acceso a atención médica y un sistema de salud eficiente. Esto resulta en una población más sana, incrementando la productividad laboral y, por ende, el GDP per capita. \
**Literacy (%):** Esta variable indica la tasa de alfabetización. Una tasa alta está relacionada con una fuerza laboral más educada, que puede adaptarse rápidamente a nuevas tecnologías y mejorar la productividad. De esta manera, se relaciona con un aumento en el GDP per capita.\
**Phones (per 1000):**Esta variable indica el acceso a teléfonos, la cual está relacionada con la mejora de la comunicación y el acceso a información. Esto facilita el comercio, el aprendizaje y la eficiencia en los negocios. En países con un alto número de teléfonos por cada 1,000 personas, se promueve la innovación y la conectividad, lo que puede indicar un mayor GDP per capita.\
**Arable (%):**Indica el porcentaje de tierra cultivable. Una mayor proporción de tierra cultivable significa más producción agrícola, lo que puede generar ingresos para el país, aumentando el GDP per cápita.\
**Crops (%):**Indica el porcentaje de cultivos. Una tasa alta en la economía refleja dependencia de la agricultura, lo que puede influir en el GDP per capita dependiendo de la eficiencia del sector. Si los cultivos no están bien gestionados o el país depende demasiado de ellos sin diversificar, el crecimiento del GDP per capita puede ser bajo. Sin embargo, si los cultivos son bien gestionados y exportados, el impacto puede ser positivo.\
**Other (%):** Para anlizar qué indica esta variable, hay que notar que la suma de arable, crops y other siempre suma 100. Por lo tanto, es posible suponer que esta variable representa el porcentaje de tierra de un país que no está clasificada como arable (tierra cultivable) ni destinada a cultivos permanentes. Esta categoría puede incluir áreas urbanas, montañosas, bosques, desiertos y otras formas de uso del suelo. Países con una alta proporción de urbanización suelen tener un GDP per cápita más alto debido a una mayor concentración de capital, educación y tecnología.\
**Climate:**El clima influye en la productividad agrícola y la calidad de vida. Un clima favorable puede aumentar la producción agrícola, lo que beneficia el GDP per capita. Sin embargo, climas extremos (sequías, inundaciones, etc.) pueden afectar negativamente la productividad, destruir infraestructuras y reducir el crecimiento económico, disminuyendo el GDP per capita.\
**Birthrate:** La tasa de natalidad se relaciona con el GDP per cápita de manera inversa. A mayor tasa, más población y por ende, el GDP per cápita decrece. \
**Deathrate:** La tasa de mortalidad se relaciona con el GDP per cápita de manera directa. A mayor tasa, menor población en el país y por ende, el GDP per cápita decrece.\
**Agriculture:** Esta variable muestra la proporción del GDP derivada de la agricultura e indica la dependencia de la economía en este sector. Una alta dependencia puede ser riesgosa si hay fluctuaciones en precios o condiciones climáticas adversas. Muchos países de bajos ingresos dependen de la agricultura, pero su productividad suele ser baja,, por lo que una alta dependencia de la agricultura tiende a estar negativamente correlacionada con el PIB per cápita\
**Industry:** Un sector industrial fuerte suele asociarse con economías más desarrolladas y diversificadas, ya que la manufactura y la infraestructura suelen generar empleos mejor remunerados y fomentar la innovación. Esto contribuye positivamente al GDP per cápita por medio de una mayor producción.\
**Service:** Un sector de servicios desarrollado indica una economía orientada hacia actividades terciarias, como finanzas, educación y turismo, que suelen correlacionarse con un mayor PIB per cápita, especialmente cuando incluye sectores de alto valor agregado como tecnología y finanzas.

### Pregunta 1.14

Detecte las observaciones outliers de las tres variables seleccionadas en la pregunta anterior. Además, impute estas observaciones si usted lo considera necesario. Justifique su decisión.

```python
#para visualizarlo antes/despúes:
variables = ['Population', 'Area (sq. mi.)', 'Phones (per 1000)']

#creamos una figura con subplots (antes), vemos los outliers
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, variable in enumerate(variables):
    df.plot.scatter(x=variable, y="GDP ($)", ax=axes[i], title=f"{variable} (Antes)")

plt.tight_layout()
plt.show()

#filtramos outliers
df2 = df[(df['Population'] < 1e9) &  # menos de 1000 millones de habitantes
         (df['GDP ($)'] < 2e12) &  # menos de 2 trillones de dólares
         (df['Area (sq. mi.)'] < 0.75e7) &  # área menor a 7.5M de millas cuadradas
         (df['Phones (per 1000)'] < 900)]  # menos de 900 teléfonos por cada 1000 habitantes

#creamos una figura con subplots (después)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, variable in enumerate(variables):
    df2.plot.scatter(x=variable, y="GDP ($)", ax=axes[i], title=f"{variable} (Después)")

plt.tight_layout()
plt.show()
```

```python
#realizo un top 5 para identificar cuáles son los países outliers en los gráficos.
top_5_gdp = df.sort_values('GDP ($)', ascending=False).head(5)
print(top_5_gdp[['Country', 'GDP ($)']])
top_5_phones = df.sort_values('Phones (per 1000)', ascending=False).head(5)
print(top_5_phones[['Country', 'Phones (per 1000)']])
top_5_pop = df.sort_values('Population', ascending=False).head(5)
print(top_5_pop[['Country', 'Population']])
top_5_area = df.sort_values('Area (sq. mi.)', ascending=False).head(5)
print(top_5_area[['Country', 'Area (sq. mi.)']])
```

Para 'Population' imputé los 3 outliers identificados que serían Estados Unidos, China e India (los dos primeros fueron eliminados filtrando por GDP). En 'Phones' eliminé un outlier correspondiente al país Monaco, que tiene en promedio 1035 teléfonos por cada 1000 habitantes, lo cual está medianamente alejado del siguiente dato con tasa más alta (Estados Unidos).
